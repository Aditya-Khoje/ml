```python
#Checking Outlier by definition and treating outliers

Airpollution_df = pd.DataFrame(temp['Air pollution (total) (deaths per 100,000)'])
Airpollution_median = Airpollution_df.median()

#getting IQR
Q3 = Airpollution_df.quantile(q=0.75)
Q1 = Airpollution_df.quantile(q=0.25)
IQR = Q3-Q1

#Deriving boundaries of Outliers
IQR_LL = int(Q1 - 1.5*IQR)
IQR_UL = int(Q3 + 1.5*IQR)

#Finding and treating outliers - both lower and upper end
temp.loc[temp['Air pollution (total) (deaths per 100,000)']>IQR_UL , 'Air pollution (total) (deaths per 100,000)'] = int(Airpollution_df.quantile(q=0.90))
temp.loc[temp['Air pollution (total) (deaths per 100,000)']<IQR_LL , 'Air pollution (total) (deaths per 100,000)'] = int(Airpollution_df.quantile(q=0.01))
```

```python
bxplt = sns.boxplot(temp['Air pollution (total) (deaths per 100,000)'])
plt.show()
```
