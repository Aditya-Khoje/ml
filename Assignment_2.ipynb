```python
#Checking Outlier by definition and treating outliers

#getting median citric acid
citric_acid_df = pd.DataFrame(MyData['citric acid'])
citric_acid_median = citric_acid_df.median()

#getting IQR of citric acid column
Q3 = citric_acid_df.quantile(q=0.75)
Q1 = citric_acid_df.quantile(q=0.25)
IQR = Q3-Q1

#Deriving boundaries of Outliers
IQR_LL = int(Q1 - 1.5*IQR)
IQR_UL = int(Q3 + 1.5*IQR)

#Finding and treating outliers - both lower and upper end
MyData.loc[MyData['citric acid']>IQR_UL , 'citric acid'] = int(citric_acid_df.quantile(q=0.90))
MyData.loc[MyData['citric acid']<IQR_LL , 'citric acid'] = int(citric_acid_df.quantile(q=0.01))
```

```python
#Check max age value now
max(MyData['citric acid'])
plt.boxplot(MyData['citric acid'])
plt.show()
```

```python
#Checking Outlier by definition and treating outliers

#getting median alcohol
alcohol_df = pd.DataFrame(MyData['alcohol'])
alcohol_median = alcohol_df.median()

#getting IQR of citric acid column
Q3 = alcohol_df.quantile(q=0.75)
Q1 = alcohol_df.quantile(q=0.25)
IQR = Q3-Q1

#Deriving boundaries of Outliers
IQR_LL = int(Q1 - 1.5*IQR)
IQR_UL = int(Q3 + 1.5*IQR)

#Finding and treating outliers - both lower and upper end
MyData.loc[MyData['alcohol']>IQR_UL , 'alcohol'] = int(alcohol_df.quantile(q=0.90))
MyData.loc[MyData['alcohol']<IQR_LL , 'alcohol'] = int(alcohol_df.quantile(q=0.01))
```

```python
#Check max age value now
max(MyData['alcohol'])
plt.boxplot(MyData['alcohol'])
plt.show()
```
